Type of data source:
This is internal database data. It comes from the company’s inventory system.

##### Possible quality issues:

- Negative stock values

- Missing warehouse location
- Outdated update dates

##### Validation before ML use:

- Make sure SKU is unique

```python
#step3
def validate_registrations(entries):
    valid_entries=[]
    invalid_entries=[]

    for entry in entries:
        try:
            username_valid=bool(entry["username"])
            email_valid="@"in entry["email"]
            age_valid=int(entry["age"])>0

            if username_valid and email_valid and age_valid:
                valid_entries.append(entry)
            else:
                invalid_entries.append(entry)

        except:
             invalid_entries.append(entry)
    return valid_entries,invalid_entries
```

#### Task 3: Row-Major vs Column-Major Access Patterns
This task lets you observe the performance differences between row-level and column-level data access empirically.

**Step 1:** Create a large DataFrame with 200,000 rows and 20 numeric columns (random float values). Name the columns feature_0 through feature_19.

**Step 2:** Measure and compare the time for these two operations:

Operation A — Column access: Compute the mean of feature_0 using the pandas column accessor (df["feature_0"].mean()).

Operation B — Row iteration: Loop through the first 10,000 rows using df.iloc[i] and sum the values of feature_0 manually.

Use %%timeit or %time to measure each. Record the results.

**Step 3:** Now convert the DataFrame to a NumPy array and repeat both operations:

Operation C — NumPy column access: Compute the mean of column 0 (arr[:, 0].mean()).

Operation D — NumPy row access: Access the first 10,000 rows sequentially (arr[i] in a loop) and sum column 0.

Record the results.

**Step 4:** Write a markdown cell that answers:

Why is column access in pandas so much faster than row iteration?
Why does the gap narrow (or change) when using NumPy?
What does this tell you about how pandas and NumPy lay out data in memory?


```python
#step1
import pandas as pd
import numpy as np

np.random.seed(42)

data = np.random.rand(200000, 20)

columns = [f"feature_{i}" for i in range(20)]

df = pd.DataFrame(data, columns=columns)

df.head()
```

```python
#step2
%timeit df["feature_0"].mean()
```

4.38 ms ± 219 μs per loop (mean ± std. dev. of 7 runs, 100 loops each)

```python
%%timeit
total = 0
for i in range(10000):
    total += df.iloc[i]["feature_0"]
```

987 ms ± 35.4 ms per loop (mean ± std. dev. of 7 runs, 1 loop each)

```python
#step3
arr=df.to_numpy()
```

```python
timeit arr[:,0].mean()
```

2 ms ± 156 μs per loop (mean ± std. dev. of 7 runs, 100 loops each)

```python
%%timeit
total=0
for i in range(10000):
    total+=arr[i,0]
```

3.53 ms ± 341 μs per loop (mean ± std. dev. of 7 runs, 100 loops each)

#step4
##### 1) Why is column access in pandas much faster than row iteration?
- Because pandas stores data column-wise.
- Column operations use vectorized functions (written in C), which are very fast.
- Row iteration uses a Python loop, and Python loops are slow.

##### 2) Why does the gap narrow when using NumPy?
- NumPy has less overhead than pandas.
- Row access is faster than in pandas.
- But if you still use a Python loop, it is slower than vectorized operations.

##### 3) What does this tell us about memory layout?
- Pandas is optimized for column-based operations.
- NumPy stores data in contiguous memory blocks.
- Vectorized operations are always faster than row-by-row loop

### Type Differences

In CSV:
- Numeric columns are usually preserved.
- String columns are stored as object.
- Sometimes type information can be lost.

In Parquet:
- All data types are preserved exactly.
- Numeric columns remain float.
- String columns remain proper string type.

Parquet preserves types better because it stores metadata.
CSV only stores plain text.

### Why can Parquet read only one column?

Parquet stores data in a column-based format.
Each column is saved separately inside the file.
So when we request only one column, Parquet loads only that column.

CSV stores data row-by-row as plain text.
To get one column, the entire file must be scanned.
That is why CSV cannot efficiently read just one column.

Conclusion:
Parquet is faster and more efficient for selective column reading.
